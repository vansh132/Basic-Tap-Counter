# **Tokenizing**

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

tokens = nltk.word_tokenize(para2)

#Word cloud
wordcloud = WordCloud(
                          background_color='white',

                          random_state=42
                         ).generate(str(tokens))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("tokens.png", dpi=900)
```

# **Removing Stop words**

```python
import nltk
from nltk.corpus import stopwords

# print(stopwords.words('english'))

stop_words = set(stopwords.words('english'))

filtered_sentence = []

for w in tokens:
    if w not in stop_words:
        filtered_sentence.append(w)

print("Actual Token Words\n")
print(tokens)
print("Length of words after removing stopwords:", len(tokens))
print("\nToken Words After Removing Stopwords\n")
print(filtered_sentence)
print("Length of words after removing stopwords:", len(filtered_sentence))
```

# **Stemming**

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

ps = PorterStemmer()

poster_sentence = []

for w in filtered_sentence:
    poster_sentence.append(ps.stem(w))
    print(w, " : ", ps.stem(w))

print("\n\n",poster_sentence)
print("\nLength or rather number of words ",len(poster_sentence))

#Word cloud for stemming
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stop_words,

                          random_state=42
                         ).generate(str(poster_sentence))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```

# **Lemmatizing**

```python
from nltk.stem import WordNetLemmatizer
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
lemmatizer = WordNetLemmatizer()

lem_sentc = []

for w in filtered_sentence:
    print(w, " : ", lemmatizer.lemmatize(w))
    lem_sentc.append(lemmatizer.lemmatize(w))

print("\nLength or rather number of words ",len(lem_sentc))

#Word cloud
wordcloud = WordCloud(
                          background_color='white',
                          max_words=100,
                           stopwords=stop_words,
                          random_state=42
                         ).generate(str(lem_sentc))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```

# **Conclusion: **


*   **Tokenization:** breaks sentences into words.
*   **Stop-words:** removes stop-words such as is, are, did, and, or
*   **Stemming:** removes prefix or suffixes from tokenized words and convert them to their base word. ie. - user-generated  :  user-gener
*   **Lemmatization:** converts word to their base word. ie - user-generated : user-generated
